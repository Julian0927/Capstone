## ***Connect to the database.***

Let us first load the SQL extension and establish a connection with the database.

```python
#(1) importamos estas dos librerias
import csv, sqlite3

#(2) creamos un objeto de tipo sqlie3
con = sqlite3.connect("my_data1.db")

#(3) usamos el objeto anterior para crear la instancia cur
cur = con.cursor()
```

```python
%sql sqlite:///my_data1.db
```

```python
#(1) importamos pandas
import pandas as pd

#(2) construimos un dataframe por medio de la direccion dada que describe un archivo de tipo csv
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_2/data/Spacex.csv")

```

```python
#help(pd.to_datetime)
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Date"].head()
```

```python
# pasamos el dataframe a sql
df.to_sql("SPACEXTBL", con, if_exists = 'replace', index = False, method = "multi")
```

## ***Task 2.***


Display 5 records where launch sites begin with the string 'CCA'.

## ***Task 6.***

List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000.

## ***Task 8.***

List the names of the booster_versions which have carried the maximum payload mass. Use a subquery.

```python
%%sql 

select "Booster_Version" as "Boosterversion" from "SPACEXTBL" where  "PAYLOAD_MASS__KG_" = (select max("PAYLOAD_MASS__KG_") from "SPACEXTBL")
```

## ***Task 9.***

List the records which will display:

* the month names

* failure landing_outcomes in drone ship 

* booster versions

* launch_site 

* the months in year 2015.


`Note:` SQLLite does not support monthnames. So you need to use substr(Date, 4, 2) as month to get the months and substr(Date,7,4)='2015' for year.

```python
df.head()
```

```python
df["Landing _Outcome"].values
```

```python
%%sql

select "Date", substr("Date",6,2) from "SPACEXTBL" 
 
```

```python
%%sql

select substr("Date",6,2) as "Month", "Landing _Outcome", "Booster_Version", "Launch_Site" from "SPACEXTBL" where (substr("Date",1,4) = "2015") and ("Landing _Outcome" = "Failure (drone ship)")
 
```
